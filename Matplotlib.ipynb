```python
file_path = '/content/gdrive/MyDrive/Colab Notebooks/Canadian_Immigration.xlsx'

df = pd.read_excel(file_path)

print(df.head())
```

```python
df
```

```python
df_pivot = df.pivot(index=['Countries', 'Continent', 'Region'], columns='Year', values='Total Immigrants')
df_pivot.reset_index(inplace=True)
print(df_pivot)
```

```python
df_pivot['Total'] = df_pivot.iloc[:, 3:].sum(axis=1)
```

```python
df_canada = df_pivot
df_canada.set_index('Countries', inplace=True)
df_canada
```

```python
df_canada.describe()
```

```python
df_canada.dtypes
```

**Area plots**

```python
years = list(map(int, range(1980, 2014)))

df_canada.sort_values(['Total'], ascending=False, inplace=True)

df_top5 = df_canada.head()
# selecting the top 5 countries with the most number of immigrants

df_top5 = df_top5[years].transpose()
# indices(columns) are plotted on the horizontal axis

print(df_top5.head())
```

```python
df_top5.plot(kind="area")
plt.title("Immigration of migrants to Canada")
plt.xlabel("Year")
plt.ylabel("Migrant")

plt.show()
# A larger shaded area might indicate a country with more immigrants overall.
# A consistently higher line throughout the graph suggests a country with a larger total number of immigrants.
```

**Histograms**

```python
df_1980 = df_canada.iloc[:,3]

count, bin_edges = np.histogram(df_1980)
df_1980.plot(kind='hist', xticks = bin_edges)

plt.title('Histogram of Immigration from 195 Countries in 1980')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

**Bar Charts**

**Box plots**

**Folium**

```python
import folium
world_map = folium.Map()
world_map
```

```python
world_map = folium.Map(
    location=[56.130, -106.35],
    zoom_start=4 )
world_map
```

```python
india_map = folium.Map(
    location=[23.5849, 84.8602] ,
    zoom_start=6,
#    tiles='Stamen Terrain'
)
```

```python
folium.Marker(location=[25.318611,82.971796], popup="Banaras").add_to(india_map)
```

```python
kanyakumari = folium.map.FeatureGroup()
kanyakumari.add_child(
    folium.features.CircleMarker(
        [8.088306,77.538452], radius=5,
        color='red', fill=True, fill_color='blue', fill_opacity=0.6
    )
)

india_map.add_child(kanyakumari)
india_map
```

```python
#Multiple markers
locations = [
    {"location": [27.036007,88.262672], "popup":"Darjeeling"},
    {"location": [29.392326,79.461242], "popup":"Nainital"},
    {"location": [19.0760,72.8777], "popup":"Mumbai"}
]

from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(india_map)

for loc in locations:
    folium.Marker(
        location=loc["location"],
        popup=loc["popup"]
    ).add_to(marker_cluster)

india_map
```

```python
!pip install requests
```

```python
import requests

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/world_countries.json'
resp = requests.get(URL)
world_geo = resp.json() #

print('GeoJSON file loaded!')
```

```python
df_canada = df_pivot
df_canada.reset_index(inplace=True)
df_canada.head()
```

```python
world_map = folium.Map(
    zoom_start = 3
)

world_map.choropleth(
    geo_data=world_geo,
    data=df_canada,
    columns=['Countries', 'Total'],
    key_on='feature.properties.name',
    fill_color='YlOrBr',
    fill_opacity=0.7,
    legend_name='Immigration to Canada'
)

world_map
```
